```python
def fibonacci(n):
    """Calculate fibonacci system in signal processing way; weighted sum of fundamental modes.
    """
    p = constants.golden
    return int(np.around(
        (p / (np.sqrt(5)    ) * np.power(p, n)) + \
        (1 / (np.sqrt(5) * p) * np.power(-p, -n))
    ))

for i in range(1000):
    print(fibonacci(i))
```
